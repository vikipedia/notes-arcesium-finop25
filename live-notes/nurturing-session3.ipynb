```python
# your code here
import time

class Timer:

    def __init__(self):
        self.startime = 0
        self.endtime = 0

    def start(self):
        self.startime = time.time()

    def stop(self):
        self.endtime = time.time()

    def get_elapsed_time(self):
        return self.endtime - self.startime
        
```

```python
%verify_problem timer
```

```python
def do_something():
    time.sleep(2)

timer = Timer()

timer.start()
do_something()
timer.stop()

print(timer.get_elapsed_time())
# 2.000
```
