````python
def vector_add(vector1, vector2):
    addition = []
    for v1, v2 in zip(vector1, vector2):
        addition.append(v1+ v2)
    return addition
````

````python
def merge(collection1, collection2):
    merged = []
    for c1, c2 in zip(collection1, collection2):
        merged.extend([c1, c2])
    return merged
````

````python
merge([1, 2, 3], ['a','b','c'])
````

````python
def factors(n):
    return [f for f in range(1, n+1) if n%f==0]
````

````python
factors(10)
````

````python
factors(5)
````

````python
def is_prime(n):
    return factors(n) == [1, n]
````

````python
is_prime(13)
````

````python
is_prime(12)
````

````python
[p for p in range(100) if is_prime(p)]
````
