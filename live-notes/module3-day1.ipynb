````python
import pandas as pd
def combined_csvs(csvfiles, outputfile):
    dfs = [pd.read_csv(f) for f in csvfiles]
    df = pd.concat(dfs)
    df.to_csv(outputfile)
````

````python
def mysum(*args):
    s = 0
    for i in args:
        s += i
    return s
````

````python
def combined_csvs(outputfile, *csvfiles): # outputfile is fixed argument, csvfiles is varaible number of arguments
    dfs = [pd.read_csv(f) for f in csvfiles]
    df = pd.concat(dfs)
    df.to_csv(outputfile)
````

````python
df1
````

````python
 df2 = pd.DataFrame(
    {
    "c":[1,2, 3, 4],
    "d":[34, 56, 76, 87],
    "labels":["x","y","z","m"]}
    )
````

````python
df2
````
